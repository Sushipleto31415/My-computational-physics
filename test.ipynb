```julia

using Plots

# Definimos la funcin que describe la EDO
function f(t, y)
    return -2y + t^2
end

# Implementamos el mtodo de Runge-Kutta de cuarto orden
function runge_kutta_4th_order(f, y0, t0, tf, h)
    n = Int((tf - t0) / h)
    t = t0
    y = y0
    ts = [t]
    ys = [y]
    
    for i in 1:n
        k1 = h * f(t, y)
        k2 = h * f(t + h/2, y + k1/2)
        k3 = h * f(t + h/2, y + k2/2)
        k4 = h * f(t + h, y + k3)
        
        y += (k1 + 2*k2 + 2*k3 + k4) / 6
        t += h
        
        push!(ts, t)
        push!(ys, y)
    end
    
    return ts, ys
end

# Parmetros iniciales
y0 = 1.0
t0 = 0.0
tf = 2.0
h = 0.1

# Resolver la EDO
ts, ys = runge_kutta_4th_order(f, y0, t0, tf, h)

# Graficar la solucin
plot(ts, ys, label="Solucin numrica (RK4)", xlabel="t", ylabel="y(t)", title="Solucin de la EDO usando Runge-Kutta de 4to orden")
```
