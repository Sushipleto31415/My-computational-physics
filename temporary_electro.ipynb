```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use("bmh")
```

```python


def capacitor_q(t,sigma,q_0):
    return q_0 * np.exp(-sigma * t)

def E(t,sigma,q_0,c,d):
    return q_0/ (c*d)* np.exp(-sigma * t)

t=np.linspace(0,10,1000)

plt.plot(t,capacitor_q(t,1,1),label="$Q(t)$",color="purple")
plt.axhline(1, alpha=0.5, label="$Q_0=1$",color="black")


plt.title("Carga de un capacitor")
plt.ylabel("$Q(t)[C]$")
plt.xlabel(" $t[s]$")
plt.legend()
plt.savefig("Carga.png")
plt.show()


plt.plot(t,E(t,1,1,1,2),label="$E(t) \quad, d=2$",color="red")
plt.axhline(1, alpha=0.5, label="$Q_0=1$",color="black")


plt.title("Campo elctrico de un capacitor")
plt.ylabel("$E(t)[V/m]$")
plt.xlabel(" $t[s]$")
plt.legend()
plt.savefig("Campo.png")
plt.show()
```
